```python
import gym 
import os
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.env_util import make_atari_env
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3 import A2C
```

```python
env = make_atari_env('SpaceInvaders-v0',n_envs=4,seed=1)
```

```python
env = VecFrameStack(env,n_stack=4)
```

```python
model = A2C('CnnPolicy',env,verbose=1)
model.learn(total_timesteps=2_00_000)
```

```python
env = make_atari_env('SpaceInvaders-v0',n_envs=1,seed=0)
env = VecFrameStack(env,n_stack=4)
```

```python
evaluate_policy(model,env,n_eval_episodes=10)
```

```python
if not os.path.exists('Models'):
  os.mkdir('Models')
```

```python
model_path = os.path.join('Models','SPACE_INVADERS_200K')
```

```python

for v in range(1,1001):
  obs = env.reset()
  total = 0
  done = False
  while done != True:
    action,_ = model.predict(obs,deterministic=True)
    obs,reward,done,info = env.step(action)
    total += reward 
  print(' EPISODE : {} REWRARD : {} '.format(v,total))
```

```python
model.save(model_path)
```
