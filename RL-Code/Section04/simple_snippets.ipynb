## Epsilon decay example run

```python
total_episodes = 100
for episode_num in range(total_episodes):
    epsilon = 1.0 / ((episode_num/10) + 1)
    print(epsilon)
```

## Numpy vstack and array

```python
li = [i for i in range(16)]
print("Original list", li)
print("np.array", np.array(li))
print("np.vstack", np.vstack(li))
```

## Selective update per row in numpy

```python
batch_size = 16
output_size = 2
output = np.random.rand(batch_size, output_size)
print(output)
```

```python
output[0, 0] = -1
print(output)
```

```python
output[[0, 1], 0] = -1
print(output)
```

```python
to_update_rows = np.arange(16)
print(to_update_rows)

to_update_cols = [i % 2 for i in range(16)]
print(to_update_cols)
```

```python
output[to_update_rows, to_update_cols] = 0.0
print(output)
```
