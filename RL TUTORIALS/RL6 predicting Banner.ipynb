```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
plt.style.use('ggplot')
```

```python
df = pd.DataFrame()
for i in range(5):
    df['banner_'+str(i)] = np.random.randint(0,2,100000)
```

```python
df.head()
```

```python
counts = np.zeros(5)
sum_rewards = np.zeros(5)
Q = np.zeros(5)
num = 100000
```

```python
def greedy_policy(epsilon=0.8):
    
    if np.random.uniform(0,1) < epsilon:
        return np.random.choice(5)
    else:
        return np.argmax(Q)
```

```python
banner_list = []
```

```python
for i in range(num):
    banner = greedy_policy(0.5)
    
    reward = df.values[i,banner]
    
    counts[banner]  += 1
    sum_rewards[banner] += reward
    
    Q[banner] = sum_rewards[banner] / counts[banner]
    
    banner_list.append(banner)
```

```python
banner_list
```

```python
print("The optimial Banner is ", np.argmax(Q)+1)
```

```python
sns.countplot(banner_list).set(xlabel="Banner",ylabel="Counts")
plt.show()
```
