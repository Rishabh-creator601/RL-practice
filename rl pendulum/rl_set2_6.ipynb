```python
import gym 
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3 import   TD3,A2C,SAC
import os
```

```python
env = gym.make('Pendulum-v0')
```

```python
model = SAC('MlpPolicy',env=env,verbose=1)
model.learn(total_timesteps=10_000)
```

```python
if not os.path.exists('Training/Models'):
  os.mkdir('Training')
  os.mkdir('Training/Models')
```

```python
model_path = os.path.join('Training','Models','Pendulum_Model_SAC')
```

```python
model.save(model_path)
```

```python
evaluate_policy(model,env,n_eval_episodes=20)
```

```python
obs = env.reset()
total = 0

for i in range(1,1001):
  done = False
  while not done:
    action,_ = model.predict(obs)
    obs,reward,done,info = env.step(action)
    total += reward 
  print('Reward : ',total)
```
